```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Selecting relevant features for clustering
features = ['Age', 'Purchase Amount', 'Customer Rating', 'Num Purchases Last Year', 'Total Spend Last Year']
X = crm_data[features]

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Using the Elbow method to find the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow method graph
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Selecting relevant features for clustering
features = ['Age', 'Purchase Amount', 'Customer Rating', 'Num Purchases Last Year', 'Total Spend Last Year']
X = crm_data[features]

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Using the Elbow method to find the optimal number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow method graph
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Applying KMeans with 4 clusters
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(X_scaled)
crm_data['Cluster'] = clusters

# Visualizing the clusters using a pairplot
sns.pairplot(crm_data, hue='Cluster', palette='viridis', vars=features, diag_kind='kde')
plt.suptitle('Clusters Visualization', y=1.02)
plt.show()
```

```python
# Applying KMeans clustering with 4 clusters
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)

# Adding cluster labels to the original dataset
crm_data['Cluster'] = cluster_labels

# Using PCA for dimensionality reduction to visualize clusters
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pc_df['Cluster'] = cluster_labels

# Visualizing the clusters
plt.figure(figsize=(10, 7))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pc_df, palette='viridis', s=60, alpha=0.7)
plt.title('Clusters of Customers')
plt.show()
```

```python
# Analyzing the average values of key features for each cluster
cluster_analysis = crm_data.groupby('Cluster')[features].mean()

# Count of customers in each cluster
cluster_counts = crm_data['Cluster'].value_counts().sort_index()
cluster_analysis['Number of Customers'] = cluster_counts

cluster_analysis
```

```python
# Assumptions
average_customer_lifespan = 3  # Assuming an average customer lifespan of 3 years

# Calculating Average Purchase Value and Purchase Frequency for each cluster
clv_data = crm_data.groupby('Cluster').agg({
    'Purchase Amount': 'mean',
    'Customer ID': 'count'
}).rename(columns={'Purchase Amount': 'Average Purchase Value', 'Customer ID': 'Number of Customers'})

# Assuming each customer made one purchase in our dataset, so the purchase frequency is 1 for our analysis
clv_data['Purchase Frequency'] = 1

# Calculating CLV
clv_data['CLV'] = clv_data['Average Purchase Value'] * clv_data['Purchase Frequency'] * average_customer_lifespan
clv_data
```

```python
import datetime as dt

# Current date (assuming today's date for the analysis)
current_date = dt.datetime.now()

# Convert 'Date of Purchase' to datetime format
crm_data['Date of Purchase'] = pd.to_datetime(crm_data['Date of Purchase'])

# Calculate Recency
crm_data['Recency'] = (current_date - crm_data['Date of Purchase']).dt.days

# Frequency and Monetary values are already available in the dataset
crm_data['Frequency'] = crm_data['Num Purchases Last Year']
crm_data['Monetary'] = crm_data['Total Spend Last Year']

# Displaying the first few rows with Recency, Frequency, and Monetary values
crm_data[['Customer ID', 'Recency', 'Frequency', 'Monetary']].head()
```

```python
# Segmenting customers based on RFM quartiles
r_labels = range(4, 0, -1)  # Recency: Lower recency is better
f_labels = range(1, 5)      # Frequency: Higher frequency is better
m_labels = range(1, 5)      # Monetary: Higher monetary value is better

# Assigning quartile values
crm_data['R_Quartile'] = pd.qcut(crm_data['Recency'], 4, labels=r_labels)
crm_data['F_Quartile'] = pd.qcut(crm_data['Frequency'], 4, labels=f_labels)
crm_data['M_Quartile'] = pd.qcut(crm_data['Monetary'], 4, labels=m_labels)

# Combining RFM quartile values to create an RFM_Segment
crm_data['RFM_Segment'] = crm_data['R_Quartile'].astype(str) + crm_data['F_Quartile'].astype(str) + crm_data['M_Quartile'].astype(str)

# Displaying the first few rows with RFM quartile values and RFM segment
crm_data[['Customer ID', 'Recency', 'R_Quartile', 'Frequency', 'F_Quartile', 'Monetary', 'M_Quartile', 'RFM_Segment']].head()
```
